```python
input_tensor = torch.ones(4,10,1,128,128)
```

```python
def _interpolate(tensor, input_size_scale, mode):
    tensor_out = []
    for t in range(tensor.size()[1]):
        tensor_t = torch.nn.functional.interpolate(tensor[:,t,:,:,:], size=input_size_scale, mode=mode)
        tensor_out.append(tensor_t)
    tensor_out = torch.stack(tensor_out, 1)
    return tensor_out
```

```python
_interpolate(input_tensor, 256.0, 'bilinear')
```
